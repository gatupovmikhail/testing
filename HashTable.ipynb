```python
from math import isnan
def hash_key_put(keys,val,key): # Возвращает индекс массива h по данному ключу key.
    # Этот индекс массива указывает на свободную ячейку памяти. Если ячейка с данным индексом
    # уже занята, возвращается индекс ячейки через q=3 клеток (если та свободна)
    h1=int(key)%len(keys)
    k=0
    q=3
    for k in range(len(keys)):
        h=(h1+k*q)%len(keys)
        if isnan(keys[h]) or isnan(val[h]):
            return h
    size = len(keys)
    for i in range(size):
        keys.append(float('nan'))
        val.append(float('nan'))
    #print(ar)
    for k in range(len(keys)):
        h=(h1+k*q)%len(keys)
        if isnan(keys[h]) or isnan(val[h]):
            return h

def put(keys,values,key,value): # Присваивание значения value по ключу key.
    index = hash_key_put(keys,values,key)
    keys[index] = key
    values[index] = value
    return index

def find(keys,values,key): # Поиск значения по ключу key
    index0 = int(key)%len(keys)
    k = 0
    q = 3
    for k in range(len(keys)):
        index = (index0+k*q)%len(keys)
        if isnan(keys[index]):
            return "Key doesn't exsist"
        if key == keys[index] and isnan(values[index]):
            return 'The value was deleted'
        if key == keys[index]:
            return values[index]

def delete(keys,values,key): # удаление значения по ключу key
    index0 = int(key)%len(keys)
    k = 0
    q = 3
    for k in range(len(keys)):
        index = (index0+k*q)%len(keys)
        if isnan(keys[index]):
            return "Key doesn't exsist"
        if key == keys[index] and isnan(values[index]):
            return 'The value already has been deleted'
        if key == keys[index]:
            values[index] = float('nan')
            return 'Value was deleted. Success'

keys = [float('nan')]*10
values = [float('nan')]*10
print('Начальное состояние')
print('Ключи ',keys)
print('Значения ',values)
print()
print('Добавляем ключ 561564564 со значением 1')
index =put(keys,values,561564564,1)
print('Ключи ',keys)
print('Значения ',values)
print()
print('Добавляем ключ 5 со значением 3')
index = put(keys,values,5,3)
print('Ключи ',keys)
print('Значения ',values)
print()
print('Добавляем ключ 15 (остаток тот же, что и у 5) со значением 4')
index = put(keys,values,15,4)
print('Ключи ',keys)
print('Значения ',values)
print()
print('Находим значение по ключу 15')
print(find(keys,values,15))
print()
print('Удаляем значение по ключу 5')
print(delete(keys,values,5))
print()
print('Опять пытаемся удалить значение по ключу 5')
print(delete(keys,values,5))
print()
print('Пытаемся найти значение по ключу 5')
print(find(keys,values,5))
print()
print('Присваиваем значение 8 по ключу 5')
put(keys,values,5,8)
print('Ключи ',keys)
print('Значения ',values)
print()
print('Находим значение по ключу 5')
print(find(keys,values,5))
print()
print('Пытаемся найти значение по ключу 555')
print(find(keys,values,555))
    
```
